### stuff document chain text summarization

In this technique we use following structure 

PDF -> 10 DOCUMENTS -> COMBINED TEXT OF 10 DOCUMENTS -> PROMPT TEMPLATE {COMBINED TEXT} -> llM -> OUTPUT 

We use only small documents because all the documents combined and added in prompt template then feeded to LLM, as we have limitation of context in LLM thats why stuff documents used for small documents.

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs 
```

```python
from langchain_core.prompts import PromptTemplate

template = """
Write a concise and short summary of the following speech,
speech : {text}
"""

prompt =PromptTemplate(input_variables=['text'],
                       template=template) 
```

```python
from langchain.chains.summarize import load_summarize_chain

chain = load_summarize_chain(llm, chain_type= "stuff", prompt = prompt, verbose= True)

output_summary = chain.run(docs)
```

#### Redine chain for summarization

in this chain have same structure as map reduce but the summarization of any indivial chunk is aggrgated to next chunk prompt 

```python
chain = load_summarize_chain(llm, chain_type= "redine", prompt = prompt, verbose= True)
chain.run(docs)
```

### Map reduce summarization

In this technique we use following structure 

 LARGE ->DOC 1 -> PROMPT-> LLM -> SUMMARIZATION 1 ->
DOC      DOC 2 -> PROMPT-> LLM -> SUMMARIZATION 2 ->
         DOC 3 -> PROMPT-> LLM -> SUMMARIZATION 3 -> -> FINAL SUMMARY
         DOC 4 -> PROMPT-> LLM -> SUMMARIZATION 4 ->
         DOC 5 -> PROMPT-> LLM -> SUMMARIZATION 5 ->

We use only small documents because all the documents combined and added in prompt template then feeded to LLM, as we have limitation of context in LLM thats why stuff documents used for small documents.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
final_docs = RecursiveCharacterTextSplitter(chunk_size = 2000, chunk_overlap = 100).split_documents(docs)
final_docs
```

```python
len(final_docs)
```

```python
chunk_prompt = """
Please summarise the below speech:
Speech:{text}
Summary:
"""

map_prompt_template = PromptTemplate(input_variables=['text'],
                                     template = chunk_prompt)
```

```python
summary_chain = load_summarize_chain(
    llm,
    chain_type = "map_reduce",
    map_prompt = map_prompt_template
)
```

```python
final_prompt = """
Provide the final summary of the entire speech with those points.
Add a Title, start the precise summary with an introduction and provide the summary in points for the specch.
Speech:{text}
"""

final_prompt_template = PromptTemplate(
    input_variables=['text'],
    template=final_prompt
)
```

```python
summary_chain = load_summarize_chain(
    llm,
    chain_type = "map_reduce",
    map_prompt = map_prompt_template,
    combine_prompt = final_prompt_template,
    verbose = True
)
```

```python
output = summary_chain.run(final_docs)
output
```
